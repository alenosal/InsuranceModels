## Analiza zbioru


Analizowany przez nas zbiór zawiera dane dotyczące ceny ubezpieczenia w zależnośći od zmiennych takich jak: cena ubezpieczenia,  wiek, płeć, ilość dzieci, to czy osoba pali, czy z jakiego regionu pochodzi. Zbiór zawiera 1338 wierszy. Celem naszego projektu jest obliczenie za pomocą regresji potencjalnej ceny ubepieczenia na podstawie podanych wartośći i sprawdzenie, które parametry nie wpływają na wysokość ceny.

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
import random

data = pd.read_csv('./insurance.csv', delimiter=';')
data.head()
```

```python
data.describe()
```

W poniższej komórce zastępujemy brakujące dane BMI średnią z całego zbioru plus jakaś losowa liczba z przedziały od -5 do 5, aby wartośći nie były dla wszystkich takie same

```python
m = data['bmi']
m = m[m!=0]
mean_bmi = np.mean(m)
data.loc[data.bmi == 0, "bmi"] = mean_bmi + random.uniform(-5.0, 5.0)
data['bmi'].describe()
```

Poniżej używamy regresji liniowej, aby uzupełnić brakujące dane dla wartości charges. do tego użyjemy wartości które są najbardzeiej skorelowane z nią czyli age, bmi i smoker.

```python
data_prep = data[data['charges'] >= 0]
X = data_prep[['age','bmi', 'smoker' ]]
y = data_prep['charges']

model = LinearRegression()
model.fit(X,y)

zeroIndexes = data.index[data['charges'] <= 0].tolist()
for i in zeroIndexes:
  data.loc[i, 'charges'] = model.predict([[data['age'].iloc[i], data['bmi'].iloc[i], data['smoker'].iloc[i]]])
data['charges'].describe()
```

```python
X = data_prep[['age', 'bmi', 'children']]

standard = preprocessing.scale(X)
scaled_features_df = pd.DataFrame(standard, index=X.index, columns=X.columns)
print(scaled_features_df)
scaled_features_df = pd.concat([scaled_features_df, 
                               data_prep[["sex", "smoker", "northwest", "southwest", "northeast", "southeast"]]], axis = 1)
scaled_features_df.to_csv("sdata.csv")
```
